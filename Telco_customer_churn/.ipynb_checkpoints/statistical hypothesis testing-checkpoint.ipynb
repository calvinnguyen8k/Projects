#### Suppose our Null Hypothesis is H0: population Churn mean = .25
#### Alternate Hypothesis H1: Population Churn mean > .25
#### We now perform ztest and ttest to prove which hypothesis is correct.

```python
# imports
import math
import numpy as np
from numpy.random import randn
from statsmodels.stats.weightstats import ztest
# Generate a random array of 80 numbers having mean df.Churn.mean() and sd df.Churn.std()/math.sqrt(80)
mean = df1.Churn.mean()
sd = df1.Churn.std()/math.sqrt(80)
alpha =0.05
null_mean =.25
data = sd*randn(80)+mean
# print mean and sd
print('mean=%.2f stdv=%.2f' % (np.mean(data), np.std(data)))

# now we perform the test. In this function, we passed data, in value parameter
# we passed mean value in the null hypothesis, in alternative hypothesis we check whether the
# mean is larger

ztest_Score, p_value= ztest(data,value = null_mean, alternative='larger')
# the function outputs a p_value and z-score corresponding to that value, we compare the
# p-value with alpha, if it is greater than alpha then we do not null hypothesis
# else we reject it.
print("p-value is:",float(p_value))
if (p_value < alpha):
    
    
    print("Reject Null Hypothesis")
    
else:
        
        
    print("Fail to Reject NUll Hypothesis")
```

Therefore Churn mean is more than 25%

## One sample t-test : The One Sample t Test determines whether the sample mean is statistically different from a known or hypothesised population mean. The One Sample t Test is a parametric test.

```python
from scipy.stats import ttest_1samp

churn_mean = np.mean(df1.Churn)
print(churn_mean)
ttest, pval = ttest_1samp(df1.Churn, .25)
print("p-value is:", pval)
if pval < 0.05:    # alpha value is 0.05 or 5%
    print("we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```
